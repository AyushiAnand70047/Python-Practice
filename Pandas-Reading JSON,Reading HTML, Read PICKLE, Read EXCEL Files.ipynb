### Read JSON to CSV

```python
import pandas as pd
import numpy as np
```

```python
Data = '{"employee_name":"Ayushi","email":"ayushianand088@gmail.com","job_profile":[{"title1":"Team Lead","title2":"Senior Developer"}]}'
```

```python
df1 = pd.read_json(Data)
df1.head()
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',header=None)
```

```python
df.head()
```

### Reading HTML Content

```python
url = 'https://www.fdic.gov/bank/individual/failed/banklist.html'
dfs = pd.read_html(url)
```

```python
dfs[0]
```

```python
url_mcc = 'https://en.wikipedia.org/wiki/Mobile_country_code'
df2 = pd.read_html(url_mcc,match='Country',header=0)
```

```python
df2[0]
```
